# **Python for Data Analytics: Day 6: Seaborn**


Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

Seaborn helps you explore and understand your data. Its plotting functions operate on dataframes and arrays containing whole datasets and internally perform the necessary semantic mapping and statistical aggregation to produce informative plots. Its dataset-oriented, declarative API lets you focus on what the different elements of your plots mean, rather than on the details of how to draw them.

We have some advantages of using seaborn in our application which is as follows;

By using the seaborn library, we can easily represent our data on a plot.
This library is used to visualize our data; we do not need to take care of the internal details; we just have to pass our data set or data inside the relplot() function, and it will calculate and place the value accordingly.
Inside this, we can switch to any other representation of data using the ‘kind’ property inside it.
It creates an interactive and informative plot to representation our data; also, this is easy for the user to understand and visualize the records on the application.
It uses static aggregation for plot generation in python.
As it is based on the matplotlib so while installing seaborn, we also have other libraries installed, out of which we have matplotlib, which also provides several features and functions to create more interactive plots in python.

**Exploring Comparisions/Trends in Data**

**Line Plots**

*Line charts are best to show trends over a period of time, and multiple lines can be used to show trends in more than one group.*

```python
sns.set_style("darkgrid")
```

```python
tips
```

```python
# Set the width and height of the figure
plt.figure(figsize=(14,6))

# Add title
plt.title("Total Bill per Day")

# Line chart showing the total bill per day'
sns.lineplot(x='day', y='total_bill', data=tips, estimator=np.sum, label='sum', ci=None) # you can change it by using the estimator

# Add label for horizontal axis
plt.legend()
```

```python
# Set the width and height of the figure
plt.figure(figsize=(14,6))

# Add title
plt.title("Total Bill per Day")

# Line chart showing the average total bill per day'
# sns.lineplot(x='day', y='total_bill', data=tips, label='mean', ci=None) # by default it shows the mean
sns.lineplot(x='day', y='total_bill', data=tips, estimator=np.mean, label='sum') # you can change it by using the estimator
# sns.lineplot(x='day', y='total_bill', data=tips, estimator=np.median, label = 'median')

# Add label for horizontal axis
plt.legend()
```

**Line Plots with multiple categories**

```python
# Set the width and height of the figure
plt.figure(figsize=(14,6))

# Add title
plt.title("Total Bill per Day")

# Line chart showing the average total bill per day'
sns.lineplot(x='day', y='total_bill', data=tips, hue='sex', ci=None, estimator=np.sum, palette='cubehelix') # hue has to be categorical


# Add label for horizontal axis
plt.xlabel("Day")
plt.ylabel('Total Bill')
plt.legend()
```

**Bar Plots**

*Are useful for comparing quantities corresponding to different groups.*

```python
# Set the width and height of the figure
plt.figure(figsize=(14,6))

# Add title
plt.title("Total Bill per Day")

# Bar chart showing the average total bill per day'
sns.barplot(y='day', x='total_bill', data=tips, estimator=np.sum, palette='magma',ci=None) # more on colour palettes here https://seaborn.pydata.org/tutorial/color_palettes.html
# You can use orient = 'h' to change the orientation. note you would have to change the x and y variables too

# Add label for horizontal axis
plt.xlabel("total bill")
plt.legend()
```

**Stacked Bar Chart**

```python
total = tips.groupby('day')['total_bill'].sum().reset_index()
total
```

```python
bar1 = sns.barplot(x="day",  y="total_bill", data=total, color='orange')
```

```python
smoker = tips[tips['smoker']=='Yes']
smoker
```

```python
# set the figure size
plt.figure(figsize=(14, 8))

# top bar -> sum all values(smoker=No and smoker=Yes) to find y position of the bars
total = tips.groupby('day')['total_bill'].sum().reset_index()

# bar chart 1 -> top bars (group of 'smoker=No')
bar1 = sns.barplot(x="day",  y="total_bill", data=total, color='orange', label='Non Smoker')

# bottom bar ->  take only smoker=Yes values from the data
smoker = tips[tips['smoker']=='Yes']

# bar chart 2 -> bottom bars (group of 'smoker=Yes')
bar2 = sns.barplot(x="day", y="total_bill", data=smoker, estimator=sum, ci=None,  color='lightblue', label='Smoker')

plt.legend()
```
